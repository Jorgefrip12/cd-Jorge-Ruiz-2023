```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def regression_lineal(X, Y):
    n = len(X)
    X_mean = np.mean(X)
    Y_mean = np.mean(Y)


    numerador = sum((X - X_mean) * (Y - Y_mean))
    denominador = sum((X - X_mean) ** 2)
    b1 = numerador / denominador
    b0 = Y_mean - b1 * X_mean

    return b0, b1

def plot_regression_line(X, Y, b0, b1):
    plt.scatter(X, Y, color='blue', label='Datos')
    plt.plot(X, b0 + b1 * X, color='red', label='Regresión lineal')
    plt.xlabel('Variable independiente (X)')
    plt.ylabel('Variable dependiente (Y)')
    plt.title('Regresión Lineal')
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
    print("Ingresa los valores de X separados por espacios:")
    X = np.array(list(map(float, input().split())))

    print("Ingresa los valores de Y separados por espacios:")
    Y = np.array(list(map(float, input().split())))

    if len(X) != len(Y):
        print("Error: La cantidad de valores en X debe ser igual a la cantidad de valores en Y.")
        return
    b0, b1 = regression_lineal(X, Y)

    print(f"La ecuación de la regresión lineal es: Y = {b0:.2f} + {b1:.2f} * X")

    plot_regression_line(X, Y, b0, b1)

if __name__ == "__main__":
    main()
```
