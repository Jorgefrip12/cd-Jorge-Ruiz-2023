```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python

def estimar_pi(num_puntos):
    dentro_circulo = 0
    puntos_circulo_x = []
    puntos_circulo_y = []
    puntos_fuera_x = []
    puntos_fuera_y = []

    for _ in range(num_puntos):
        x = random.uniform(0, 1)  # Genera una coordenada x aleatoria entre 0 y 1
        y = random.uniform(0, 1)  # Genera una coordenada y aleatoria entre 0 y 1
        distancia_al_centro = x**2 + y**2  # Distancia al centro (0,0)

        if distancia_al_centro <= 1:  # Comprueba si el punto est dentro del crculo
            dentro_circulo += 1
            puntos_circulo_x.append(x)
            puntos_circulo_y.append(y)
        else:
            puntos_fuera_x.append(x)
            puntos_fuera_y.append(y)

    estimacion_pi = (dentro_circulo / num_puntos) * 4
    return estimacion_pi, puntos_circulo_x, puntos_circulo_y, puntos_fuera_x, puntos_fuera_y

num_puntos = 1000  # Puedes ajustar este nmero para obtener una mejor estimacin
valor_estimado_pi, circulo_x, circulo_y, fuera_x, fuera_y = estimar_pi(num_puntos)

# Graficar los puntos dentro y fuera del crculo
plt.figure(figsize=(6, 6))
plt.scatter(circulo_x, circulo_y, color='blue', marker='.', label='Dentro del Crculo')
plt.scatter(fuera_x, fuera_y, color='red', marker='.', label='Fuera del Crculo')

# Agregar la lnea que representa /4
x_line = np.linspace(0, 1, 100)
y_line = x_line
plt.plot(x_line, y_line, color='green', linestyle='--', label='/4')

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title(f'Estimacin de  con {num_puntos} puntos aleatorios: {valor_estimado_pi}')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

```

```python

# Genera 1000 puntos en una columna (en el eje X)
columna_x = np.random.rand(1000)

# Genera 1000 puntos en una fila (en el eje Y)
fila_y = np.random.rand(1000)

# Calcula la distancia al centro (0,0) para cada punto
distancias = np.sqrt(columna_x**2 + fila_y**2)

# Cuenta cuntos puntos estn dentro del crculo (distancia <= 1)
puntos_dentro = np.sum(distancias <= 1)

# Calcula la estimacin de  utilizando el mtodo de Monte Carlo
valor_estimado_pi = (puntos_dentro / 1000) * 4

# Graficar los puntos dentro y fuera del crculo
plt.figure(figsize=(6, 6))
plt.scatter(columna_x, fila_y, color='blue', marker='.', label='Puntos')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title(f'Estimacin de : {valor_estimado_pi}')
plt.gca().set_aspect('equal', adjustable='box')

# Dibujar el crculo
circle = plt.Circle((0, 0), 1, color='red', alpha=0.5, label='Crculo')
plt.gca().add_patch(circle)

plt.legend()
plt.show()

print(f"Valor estimado de  con 1000 puntos en una columna y 1000 puntos en una fila: {valor_estimado_pi}")

```
