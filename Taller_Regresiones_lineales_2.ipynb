```python
# Selecciona las columnas de características excluyendo "Price"
feature_columns = df.columns.difference(["Price"])

best_r_squared = -1
best_features = None

# Itera sobre todas las combinaciones posibles de 3 columnas de características
for feature_combination in combinations(feature_columns, 3):
    x = df[list(feature_combination)]
    x = sm.add_constant(x)
    y = df["Price"]

    # Transformación de variables: aplica logaritmo solo a columnas numéricas positivas
    for col in x.columns:
        if x[col].dtype == "float64" and (x[col] > 0).all():
            x[col] = np.log1p(x[col])

    model = sm.OLS(y, x)
    results = model.fit()

    r_squared = results.rsquared

    if r_squared > best_r_squared:
        best_r_squared = r_squared
        best_features = feature_combination

print("Mejor combinación de columnas:", best_features)
print("Mejor R cuadrado:", best_r_squared)
```
